```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import numpy as np

df = pd.read_csv("patient_visits.csv")

# Preview
print(df.head())

# Check missing values
print(df.isnull().sum())

# Plot trends
df['Date'] = pd.to_datetime(df['Date'])
df.groupby('Date')['NumPatients'].sum().plot(title="Daily Patient Visits", figsize=(10, 4))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import mlflow
import numpy as np

# Convert categories
df_encoded = pd.get_dummies(df, columns=['Department', 'TimeSlot', 'DayOfWeek'])

X = df_encoded.drop(['NumPatients', 'Date'], axis=1)
y = df_encoded['NumPatients']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Track with MLflow

mlflow.end_run()
mlflow.start_run()

model = RandomForestRegressor()
model.fit(X_train, y_train)

preds = model.predict(X_test)
preds = np.ravel(preds)  # Flatten just in case

# Also make sure y_test is converted to a NumPy array
y_test = np.ravel(y_test)

preds = model.predict(X_test)
def mean_squared_error(
    y_true,
    y_pred,
    *,
    sample_weight=None,
    multioutput="uniform_average",
):
 try:
    preds = model.predict(X_test)
    rmse = mean_squared_error(y_test, preds, squared=False)
    print("RMSE:", rmse)
 except Exception as e:
    print("Error during prediction or RMSE calculation:", e)
 mlflow.sklearn.log_model(model, "rf_model")

mlflow.end_run()
```

```python
import gradio as gr
import joblib
import pandas as pd

model = joblib.load("../models/rf_model.pkl")

def predict(day, department, time_slot):
    sample = pd.DataFrame([[day, department, time_slot]],
                          columns=['DayOfWeek', 'Department', 'TimeSlot'])
    sample = pd.get_dummies(sample).reindex(columns=model.feature_names_in_, fill_value=0)
    return int(model.predict(sample)[0])

gr.Interface(
    fn=predict,
    inputs=["text", "text", "text"],
    outputs="number",
    title="Patient Visit Predictor"
).launch()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("patient_visits.csv")

# Group by DayOfWeek
day_visits = df.groupby("DayOfWeek")["NumPatients"].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=day_visits.index, y=day_visits.values, palette="viridis")
plt.title("Total Patient Visits by Day of Week")
plt.ylabel("Number of Patients")
plt.xlabel("Day of Week")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by Department
dept_visits = df.groupby("Department")["NumPatients"].sum().sort_values(ascending=False)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x=dept_visits.index, y=dept_visits.values, palette="coolwarm")
plt.title("Total Patient Visits by Department")
plt.ylabel("Number of Patients")
plt.xlabel("Department")
plt.tight_layout()
plt.show()
```

```python
pivot = df.pivot_table(values='NumPatients', index='DayOfWeek', columns='Department', aggfunc='sum')

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Patient Visits by Day and Department")
plt.xlabel("Department")
plt.ylabel("Day of Week")
plt.tight_layout()
plt.show()
```
